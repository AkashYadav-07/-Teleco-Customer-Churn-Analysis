#converting the seniorCitizen value from 0 and 1 to Yes and No which can easily understand 

```python
def conv(value):
    if (value == 1):
        return "Yes"

    else:
        return "No"

df1['SeniorCitizen'] = df1['SeniorCitizen'].apply(conv)
df1.head(40)
       
```

```python
ax=sns.countplot(x ='Churn', data = df1)

ax.bar_label(ax.containers[0])
plt.title("Count the Customers By Churns")
plt.show()
```

```python
count = df1["Churn"].value_counts()
count

plt.pie(count, labels = count.index, autopct= "%1.1f")
plt.title("Percentage Of Customes Churn")
plt.show()
```

```python
plt.pie(count, labels = count.index, autopct= "%1.1f",explode= [0,0.2], shadow = True)
plt.title("Percentage Of Customes Churn")
plt.show()
```

# from the given piechart we can conclude that 26.5% of people are churned out. Now, let's explore the reason behind it.

```python
plt.figure(figsize=(5,5))

sns.countplot(x = 'gender', data = df1, palette=['Orange','Blue'], hue=df1['Churn'])
plt.title('Distribution Based on Gender ')
plt.show()
```

```python
ax = sns.countplot(x = 'SeniorCitizen', data = df1)
ax.bar_label(ax.containers[0])
plt.title('Distribution Based on SeniorCitizen ')
plt.show()
```

```python
sns.countplot(x = 'SeniorCitizen', data = df1, palette=['Orange','Blue'], hue=df1['Churn'])
plt.title('Distribution Based on SeniorCitizen ')
plt.show()
```

```python


# Percentage table
ct = pd.crosstab(df1['SeniorCitizen'], df1['Churn'], normalize='index') * 100

#  Plot stacked bar chart
ax = ct.plot(kind='bar', stacked=True, color=['orange', 'blue'])

# Add % labels inside bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# Labels and title
plt.title('Distribution Based on SeniorCitizen (%)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')

plt.show()
```
